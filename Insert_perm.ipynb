```python
import math as mt
import numpy as np
import itertools as it
import matplotlib.pyplot as plt

def randomPerm(n):
    v = np.arange(1,n,1)
    matPerm = (list(it.permutations(v)))
    return matPerm

def isortSteps(a):
    v = []
    for i in range(len(a)):
        v.append(a[i])
        
    steps = 0
    for i in range(1,len(v)):
        x = v[i]
        j = i-1
        while (j > -1) and (v[j] > x):
            v[j+1] = v[j]
            j = j -1
            steps = steps + 3
        steps = steps + 1
        v[j+1] = x
        steps = steps + 4
    steps = steps + 1
    return steps
allSteps=[]
n = 10
runs = mt.factorial(n-1)
sum = 0
min = n ** 3
max = 0
mPerm = randomPerm(n);
for i in range(runs):
    t = isortSteps(mPerm.pop())
    allSteps.append(t)
    if t < min :
        min = t
    if t > max :
        max = t
    sum = sum + t
print ('Theoretical best time, ' + str(5*n - 4) )
print ('Theoretical worst time,' + str((3.0/2.0)*n**2 + (7.0/2.0)*n - 4))
print ('Theoretical average time,' + str((3.0/4.0)*n**2 + (17.0/4.0)*n - 4))
print ('Experimenal best time, ' + str(min))
print ('Experimenal worst time,' + str(max))
print ('Experimenal average time,' + str(sum/runs))
```

```python
plt.plot(allSteps,'.')
plt.ylabel('intercambios')
plt.xlabel('n')
plt.show()
```
